```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# fichier csv contient les données des images de format 64×64

```python
# fichier csv contient les données des images de format 64×64
df=pd.read_csv("face_data.csv")
```

```python
df.head()
```

```python
def show_orignal_images(pixels):
#Displaying Orignal Images
  fig, axes = plt.subplots(10, 10, figsize=(11, 7), subplot_kw={'xticks':[], 'yticks':[]})
  for i, ax in enumerate(axes.flat):
    ax.imshow(np.array(pixels)[i].reshape(64, 64), cmap='gray')
  plt.show()
```

```python
lebels=df["target"]
lebels
```

```python
pixels=df.drop(['target'],axis = 1)
```

```python
pixels
```

```python
show_orignal_images(pixels)
```

```python
from sklearn.model_selection import train_test_split #for split data to train and test data
```

```python
x_train,x_test,y_train,y_test=train_test_split (pixels,lebels)#for split data to train and test data
```

```python
#PCA step
from sklearn.decomposition import PCA
pca=PCA(135).fit(x_train)
```

```python
per_var = np.round(pca.explained_variance_ratio_*100, decimals=1)
labels = [str(x) for x in range(1,len(per_var)+1)]
 
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.show()
```

```python
#show eigen faces
def show_eigenfaces(pca):
   #Displaying Eigenfaces
   fig, axes = plt.subplots(3, 8, figsize=(9, 4),subplot_kw={'xticks':[], 'yticks':[]})
   for i, ax in enumerate(axes.flat):
       ax.imshow(pca.components_[i].reshape(64, 64), cmap='gray')
       ax.set_title("PC " + str(i+1))
   plt.show()
```

```python
show_eigenfaces(pca)
```

```python
x_train_pca=pca.transform(x_train)
```

```python
#create_classifer
from time import time
from sklearn.metrics import classification_report
from sklearn.svm import SVC
```

```python
clf=SVC(kernel=
        "rbf",C=1000,gamma=0.01)
clf.fit(x_train_pca,y_train)
```

```python
#testing
x_test_pca=pca.transform(x_test)
y_pred=clf.predict(x_test_pca)
```

```python
print(classification_report(y_test,y_pred))
```
